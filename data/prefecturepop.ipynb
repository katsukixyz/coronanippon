```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy.types import String, Integer
```

```python
df = pd.read_excel('000701583.xls', header = 1)
```

```python
pref = df.loc[(df['団体コード'].notnull()) & (df['市区町村名'].isnull()) & (df['性別'] == '計')]
```

```python
pref = pref.reset_index(drop = True).reset_index()
```

```python
final = pref[['index', '都道府県名', '総数']]
```

```python
final['都道府県名'] = final['都道府県名'].str.replace('*', "", regex = False)
```

```python
final["index"] = final["index"].apply(lambda x: x+1)
```

```python
final = final.rename(columns = {"index": "id", "都道府県名": "name", "総数": "population"})
```

```python
final
```

```python
alchemyEngine = create_engine('postgresql+psycopg2://root:root@localhost:5432/coronanippon_db', pool_recycle=3600)
```

```python
dbConnection = alchemyEngine.connect()
```

```python
final.to_sql('prefectures', con = dbConnection, index = False, if_exists = 'replace', dtype = {"id": Integer(), "name": String(), "population": Integer()})
```
